```python
import os
import numpy as np
import pandas as pd
import seaborn as sns

# EDA for CIC-IDS2017 Friday-WorkingHours-Afternoon-DDos dataset

import matplotlib.pyplot as plt

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 5)

file_path = "data/cic-ids-2017/TrafficLabelling /Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv"
assert os.path.exists(file_path), f"File not found: {file_path}"

# Load data
df = pd.read_csv(file_path, low_memory=False)

# Clean column names and common string artifacts
df.columns = df.columns.str.strip()
for c in df.select_dtypes(include="object").columns:
    df[c] = df[c].astype(str).str.strip()

print("Shape:", df.shape)
print("\nColumns (first 15):", df.columns[:15].tolist())
print("\nData types:")
print(df.dtypes.value_counts())

print("\nHead:")
display(df.head())

# Basic quality checks
print("\nDuplicate rows:", df.duplicated().sum())

missing = df.isna().sum().sort_values(ascending=False)
missing = missing[missing > 0]
print("\nMissing values (top 20):")
display(missing.head(20) if len(missing) else pd.Series(dtype=int, name="No missing values"))

# Handle infinite values in numeric columns
num_cols = df.select_dtypes(include=[np.number]).columns
df[num_cols] = df[num_cols].replace([np.inf, -np.inf], np.nan)

# Summary stats
print("\nNumeric summary:")
display(df[num_cols].describe(percentiles=[0.01, 0.05, 0.5, 0.95, 0.99]).T)

# Target/label analysis
label_col = None
for candidate in ["Label", "label", "Class", "class"]:
    if candidate in df.columns:
        label_col = candidate
        break

if label_col is not None:
    print(f"\nLabel column detected: {label_col}")
    label_counts = df[label_col].value_counts(dropna=False)
    print("\nLabel distribution:")
    display(label_counts.to_frame("count"))
    print("\nLabel distribution (%):")
    display((label_counts / len(df) * 100).round(3).to_frame("percent"))

    plt.figure(figsize=(8, 4))
    sns.countplot(data=df, x=label_col, order=label_counts.index, palette="Set2")
    plt.title("Class Distribution")
    plt.xticks(rotation=30, ha="right")
    plt.tight_layout()
    plt.show()

# Correlation analysis (top varying numeric features to keep plot readable)
if len(num_cols) > 1:
    variances = df[num_cols].var(numeric_only=True).sort_values(ascending=False)
    top_cols = variances.head(20).index.tolist()

    corr = df[top_cols].corr(numeric_only=True)
    plt.figure(figsize=(12, 9))
    sns.heatmap(corr, cmap="coolwarm", center=0, square=False)
    plt.title("Correlation Heatmap (Top 20 High-Variance Numeric Features)")
    plt.tight_layout()
    plt.show()

    # Strong pairwise correlations
    corr_abs = corr.abs()
    upper = corr_abs.where(np.triu(np.ones(corr_abs.shape), k=1).astype(bool))
    strong_pairs = (
        upper.stack()
        .sort_values(ascending=False)
        .reset_index()
        .rename(columns={"level_0": "feature_1", "level_1": "feature_2", 0: "abs_corr"})
    )
    print("\nTop 15 strongest absolute correlations:")
    display(strong_pairs.head(15))

# Distribution plots for a few key numeric columns
plot_cols = df[num_cols].nunique().sort_values(ascending=False).head(6).index.tolist()
if plot_cols:
    fig, axes = plt.subplots(len(plot_cols), 1, figsize=(10, 3 * len(plot_cols)))
    if len(plot_cols) == 1:
        axes = [axes]
    for ax, col in zip(axes, plot_cols):
        sns.histplot(df[col].dropna(), bins=50, kde=False, ax=ax, color="#4C72B0")
        ax.set_title(f"Distribution: {col}")
    plt.tight_layout()
    plt.show()

# Outlier quick check using IQR rule
outlier_report = {}
for col in num_cols:
    s = df[col].dropna()
    if s.empty:
        continue
    q1, q3 = s.quantile(0.25), s.quantile(0.75)
    iqr = q3 - q1
    if iqr == 0:
        outlier_report[col] = 0
    else:
        outliers = ((s < (q1 - 1.5 * iqr)) | (s > (q3 + 1.5 * iqr))).sum()
        outlier_report[col] = int(outliers)

outlier_df = pd.Series(outlier_report).sort_values(ascending=False).to_frame("outlier_count")
print("\nTop 15 columns by outlier count (IQR method):")
display(outlier_df.head(15))

print("\nEDA complete.")
```
